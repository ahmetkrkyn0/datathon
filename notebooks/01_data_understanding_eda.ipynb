```python
turkish_stopwords = [
    "ve", "bir", "bu", "ile", "için", "daha", "çok", "az", "de", "da",
    "ama", "fakat", "ancak", "olarak", "olan", "olduğu", "olabilir",
    "onu", "onun", "şu", "o", "her", "gibi", "kadar", "mi", "mı",
    "mu", "mü", "ise", "en", "çok", "bazı", "hem", "ya", "veya"
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

target = "career_success_score"

high_success = train[train[target] >= train[target].quantile(0.90)]
low_success = train[train[target] <= train[target].quantile(0.10)]

high_text = high_success["mentor_feedback_text"].fillna("")
low_text = low_success["mentor_feedback_text"].fillna("")
```

```python
vectorizer_high = CountVectorizer(
    max_features=30,
    stop_words=turkish_stopwords,
    ngram_range=(1, 1)
)

high_matrix = vectorizer_high.fit_transform(high_text)

high_words = pd.DataFrame({
    "word": vectorizer_high.get_feature_names_out(),
    "high_count": high_matrix.sum(axis=0).A1
}).sort_values("high_count", ascending=False)

high_words.head(20)
```

```python
vectorizer_low = CountVectorizer(
    max_features=30,
    stop_words=turkish_stopwords,
    ngram_range=(1, 1)
)

low_matrix = vectorizer_low.fit_transform(low_text)

low_words = pd.DataFrame({
    "word": vectorizer_low.get_feature_names_out(),
    "low_count": low_matrix.sum(axis=0).A1
}).sort_values("low_count", ascending=False)

low_words.head(20)
```

```python
combined_text = pd.concat([high_text, low_text])

vectorizer = CountVectorizer(
    stop_words=turkish_stopwords,
    min_df=5
)

vectorizer.fit(combined_text)

high_matrix = vectorizer.transform(high_text)
low_matrix = vectorizer.transform(low_text)

words = vectorizer.get_feature_names_out()

high_counts = high_matrix.sum(axis=0).A1
low_counts = low_matrix.sum(axis=0).A1

word_compare = pd.DataFrame({
    "word": words,
    "high_count": high_counts,
    "low_count": low_counts
})

word_compare.head()
```

```python
word_compare["high_freq"] = word_compare["high_count"] / word_compare["high_count"].sum()
word_compare["low_freq"] = word_compare["low_count"] / word_compare["low_count"].sum()

word_compare["high_low_ratio"] = (word_compare["high_freq"] + 1e-6) / (word_compare["low_freq"] + 1e-6)

word_compare.sort_values("high_low_ratio", ascending=False).head(20)
```

```python
word_compare.sort_values("high_low_ratio", ascending=True).head(20)
```

```python
bigram_vectorizer = CountVectorizer(
    stop_words=turkish_stopwords,
    ngram_range=(2, 2),
    min_df=5
)

bigram_vectorizer.fit(combined_text)

high_bigram_matrix = bigram_vectorizer.transform(high_text)
low_bigram_matrix = bigram_vectorizer.transform(low_text)

bigrams = bigram_vectorizer.get_feature_names_out()

bigram_compare = pd.DataFrame({
    "phrase": bigrams,
    "high_count": high_bigram_matrix.sum(axis=0).A1,
    "low_count": low_bigram_matrix.sum(axis=0).A1
})

bigram_compare["high_freq"] = bigram_compare["high_count"] / bigram_compare["high_count"].sum()
bigram_compare["low_freq"] = bigram_compare["low_count"] / bigram_compare["low_count"].sum()

bigram_compare["high_low_ratio"] = (
    bigram_compare["high_freq"] + 1e-6
) / (
    bigram_compare["low_freq"] + 1e-6
)

bigram_compare.sort_values("high_low_ratio", ascending=False).head(20)
```

```python
bigram_compare.sort_values("high_low_ratio", ascending=True).head(20)
```

Mentor yorumları yüksek ve düşük başarı grupları için ayrı ayrı incelendi. İlk analizde sık geçen kelimeler listelendiğinde "ve", "bir", "bu", "ile" gibi Türkçe stop word kelimelerinin öne çıktığı görüldü. Bu kelimeler anlamlı bir ayrım sağlamadığı için stop word temizliği uygulanması gerektiği değerlendirildi.

Temizlenmiş kelime analizinde yüksek başarı grubunda "güçlü", "yüksek", "başarı", "etkileyici", "mükemmel" gibi daha olumlu ifadelerin öne çıktığı; düşük başarı grubunda ise "ancak", "geliştirme", "dikkat", "problem" gibi gelişim ihtiyacını ima eden kelimelerin daha sık görüldüğü gözlemlendi.

Bu sonuç, mentor_feedback_text alanının hedef değişken için bilgi taşıyabileceğini göstermektedir. Bu nedenle modelleme aşamasında bu metin alanının TF-IDF gibi yöntemlerle modele dahil edilmesi mantıklıdır.

```python
bigram_compare["total_count"] = bigram_compare["high_count"] + bigram_compare["low_count"]

bigram_compare_filtered = bigram_compare[bigram_compare["total_count"] >= 15]

bigram_compare_filtered.sort_values("high_low_ratio", ascending=False).head(20)
```
